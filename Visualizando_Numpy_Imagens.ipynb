```python
# Exibe a imagem colorida .jpg

imagem_bgr = cv2.imread('/content/computer-vision-companies.jpg')
img = cv2.cvtColor(imagem_bgr, cv2.COLOR_BGR2RGB) #converte para ordem RGB

print(img.size) #dimens√µesem pixel da imagem
print(img.shape)
print(type(img))
```

```python
#imag_arr = np.asarray(img)
plt.imshow(img)
```

```python
#acessando cada uma das camadas (valores num√©ricos representam os pixels)

img[200:250,190:210,2]
```

```python
# Alterando a cor em um intervalo na camada 1

img[100:150,200:300,1] = 180
plt.imshow(img)
```

```python
# acessando cada cor separadamente

cor = img[:, :, 0] # todas as linhas, todas as colunas, camada 0 - R, 1 - G e 2 - B

plt.imshow(cor, cmap = 'gray') # a √°rea branca representa a cor selecionada - camada
```
